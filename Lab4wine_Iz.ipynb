```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

wine_dataset = load_wine()  # Загружаем датасет

X_train, X_test, y_train, y_test = train_test_split(wine_dataset['data'], wine_dataset['target'], random_state=0)
```

```python
print("форма массива X-train : {}".format(X_train.shape))
print("форма массива Y-train : {}".format(y_train.shape))
print("форма массива X-test : {}".format(X_test.shape))
print("форма массива Y-test : {}".format(y_test.shape))
```

```python
import pandas as pd
wine_dataframe = pd.DataFrame(X_train, columns=wine_dataset.feature_names)
grr =  pd.plotting.scatter_matrix(wine_dataframe, c=y_train, figsize=(15,15),marker='o',hist_kwds={'bins':20}, s=60, alpha=.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
import numpy as np
X_new = np.array([[1.444e-01, 6.111e+00, 2.491e+00, 5.522e+01, 6.880e-02, 2.809e+00, 3.000e+00,
  2.111e-01, 2.290e-04, 5.666e+00, 3.040e+00, 2.933e+00, 2.00e-03]])
print("форма массива X_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(wine_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("прогнозы для тестового набора:\n {}".format(y_pred))
```

```python
print("правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```
