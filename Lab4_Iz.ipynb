```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris_dataset = load_iris()  # Загружаем датасет

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print("форма массива X-train : {}".format(X_train.shape))
print("форма массива Y-train : {}".format(y_train.shape))
```

```python
print("форма массива X-test : {}".format(X_test.shape))
print("форма массива Y-test : {}".format(y_test.shape))
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
grr =  pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15),marker='o',hist_kwds={'bins':20}, s=60, alpha=.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
import numpy as np
X_new = np.array([[5, 2.9, 1, 0.2]])
print("форма массива X_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("прогнозы для тестового набора:\n {}".format(y_pred))
```

```python
print("правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'],iris_dataset['target'], random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```

Было 10 возможных ответов - 0-9 . Это задача многоклассовой классификации. В этом решении используется алгоритм классификации k ближащих соседей. 

Правильность этой модели для тестового набора составляет около 0.99, что означает что был использован правильный прогноз для 99% рукописных цифр в тестовом наборе.
