```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

breast_cancer_dataset = load_breast_cancer()  # Загружаем датасет

X_train, X_test, y_train, y_test = train_test_split(breast_cancer_dataset['data'], breast_cancer_dataset['target'], random_state=0)
```

```python
print("форма массива X-train : {}".format(X_train.shape))
print("форма массива Y-train : {}".format(y_train.shape))
print("форма массива X-test : {}".format(X_test.shape))
print("форма массива Y-test : {}".format(y_test.shape))
```

```python
y_pred = knn.predict(X_test)
print("прогнозы для тестового набора:\n {}".format(y_pred))
```

```python
print("правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
breast_cancer_dataset['data'],breast_cancer_dataset['target'], random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```

Было 2 возможнх состояния опухоли - Malignant и Benign. Это задача двуклассовой классификации. В этом решении используется алгоритм классификации k ближащих соседей. 

Правильность этой модели для тестового набора составляет около 0.92, что означает что был использован правильный прогноз для 92% пациентов в тестовом наборе. 

При параметре KneighborsClassifier(количество соседей) равном 1, правильность модели на тестовом наборе 0.92

При параметре KneighborsClassifier(количество соседей) равном 2, правильность модели в первых двух случаях на тестовом наборе 0.90
